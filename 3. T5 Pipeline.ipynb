```python
abstract_pattern = r'^(I{0,3}|[0-9]+)[ \.]*(?:A\s*[bB]\s*[sS]\s*[tT]\s*[rR]\s*[aA]\s*[cC]\s*[tT])'
introduction_pattern = r'^(I{0,3}|[0-9]+)[ \.]*(?:I\s*[nN]\s*[tT]\s*[rR]\s*[oO]\s*[dD]\s*[uU]\s*[cC]\s*[tT]\s*[iI]\s*[oO]\s*[nN])'
introduction_end_pattern = r'^((I{1,3})|[0-9]+)[ \.]+(?:[A-Z]\s*[A-Za-z ]*)'
conclusion_pattern = r'^((IX|IV|V?I{0,3})|[0-9]+)[ \.]*(?:C\s*[oO]\s*[nN]\s*[cC]\s*[lL]\s*[uU]\s*[sS]\s*[iI]\s*[oO]\s*[nN])'
reference_pattern = r'^((IX|IV|V?I{0,3})|[0-9]+)[ \.]*(?:R\s*[eE]\s*[fF]\s*[eE]\s*[rR]\s*[eE]\s*[nN]\s*[cC]\s*[eE]\s*[sS ])'
```

```python
def get_bibtex_from_doi(doi):
    api_url = f"https://api.crossref.org/works/{doi}"
    try:
        response = requests.get(api_url)
        response.raise_for_status()
        data = response.json()
        bibtex_entry = data['message']
        return bibtex_entry
    except Exception as e:
        print(f"Error: {e}")
        return None

def extract_title_and_author(bibtex_entry):
    title = bibtex_entry.get('title', 'Title not found')
    title = title[0]

    authors = bibtex_entry.get('author', 'Authors not Found')
    first_author = None
    for author in authors:
        if author['sequence'] == 'first':
            first_author = author['given'] + ' ' + author['family']

    return title, first_author
```

```python
def extract_sections(pdf_path, start_pattern, end_pattern=None):
    pdf_reader = PyPDF2.PdfReader(pdf_path)
    section_text = ""
    section_match = False

    for page in pdf_reader.pages:
        page_text = page.extract_text()
        lines = page_text.splitlines()
        for line in lines:
            if line:
                if section_match and end_pattern and re.match(end_pattern, line):
                    return section_text
                if re.search(start_pattern, line):
                    section_match = True
                if section_match:
                    section_text += line + ' ' if line[-1] != '-' else line[:-1]

    return section_text

def extract(pdf_path):
    abstract_text = extract_sections(pdf_path, abstract_pattern, introduction_pattern)
    introduction_text = extract_sections(pdf_path, introduction_pattern, introduction_end_pattern)
    conclusion_text = extract_sections(pdf_path, conclusion_pattern, reference_pattern)

    return abstract_text, introduction_text, conclusion_text
```

```python
doi = "10.1109/INOCON50539.2020.9298232"
path = "/content/drive/MyDrive/THESIS/AIC/1.pdf"

bibtex_entry = get_bibtex_from_doi(doi)
abstract, introduction, conclusion = extract(path)
if bibtex_entry:
    title, first_author = extract_title_and_author(bibtex_entry)

title1 = title
first_author1 = first_author
abstract1 = abstract
introduction1 = introduction
conclusion1 = conclusion
print(f'Title: {title1}')
print(f'First Author: {first_author1}')
print(abstract1)
print(introduction1)
print(conclusion1)
```

```python
from simplet5 import SimpleT5

model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
```

```python
model.load_model("t5","/content/drive/MyDrive/THESIS/AIC/simplet5-epoch-5-train-loss-1.3971-val-loss-3.1449", use_gpu=True)
```

```python
def main_function(serial,title,author,abstract,introduction,conclusion):

  total=abstract+introduction+conclusion
  t5_pred_tot=model.predict(total)


  summary=""
  flag=0
  for i in t5_pred_tot[0].split():
    if i=="presented." or i=="presented" or i=="proposed." or i=="proposed" or i=="propose" or i=="this":
      flag=1

    if i=="We" or i=="we":
      pass
    else:
      summary=summary+i+" "


  author_ieee=author.split()[-1]+" et al."


  if flag==0:
    starting="In the article ["+str(serial)+"], "+author_ieee+" presented "
  else:
    starting="In the article presented by "+author_ieee+" ["+str(serial)+"], "


  final=starting+summary

  final2=""

  for i in final.split():
    if i!="Abstract":
      final2=final2+i+" "




  return final2
```
