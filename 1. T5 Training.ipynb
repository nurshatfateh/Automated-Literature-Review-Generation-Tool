```python
import evaluate
rouge = evaluate.load('rouge')
```

```python
from datasets import load_dataset
ds = load_dataset("allenai/scitldr", name="AIC")
```

```python
ds
```

```python
print()
print(f"Train dataset size: {len(ds['train'])}")
print()
print(f"Test dataset size: {len(ds['test'])}")
print()
print(f"Validation dataset size: {len(ds['validation'])}")
print()
print(f"Source example: {ds['train']['source'][0]}")
print()
print(f"Target example: {ds['train']['target'][0]}")
print()
```

```python
import pandas as pd
dftrain = pd.DataFrame(ds["train"])
dftrain.head()
```

```python
dftest = pd.DataFrame(ds["test"])
dftest.head()
```

```python
dfval = pd.DataFrame(ds["validation"])
dfval.head()
```

```python
dftrain = dftrain.rename(columns={"target":"target_text", "source":"source_text"})
dftrain = dftrain[['source_text', 'target_text']]
dftrain.head()
```

```python
dftest = dftest.rename(columns={"target":"target_text", "source":"source_text"})
dftest = dftest[['source_text', 'target_text']]
dftest.head()
```

```python
dfval = dfval.rename(columns={"target":"target_text", "source":"source_text"})
dfval = dfval[['source_text', 'target_text']]
dfval.head()
```

```python
dftrain['source_text'] = dftrain['source_text'].apply(lambda x: ' '.join(x))
dftrain['target_text'] = dftrain['target_text'].apply(lambda x: ' '.join(x))
dftrain
```

```python
dftest['source_text'] = dftest['source_text'].apply(lambda x: ' '.join(x))
dftest['target_text'] = dftest['target_text'].apply(lambda x: ' '.join(x))
dftest
```

```python
dfval['source_text'] = dfval['source_text'].apply(lambda x: ' '.join(x))
dfval['target_text'] = dfval['target_text'].apply(lambda x: ' '.join(x))
dfval
```

```python
dftrain['source_text'] = "summarize: " + dftrain['source_text']
dftrain.head()
```

```python
dftest['source_text'] = "summarize: " + dftest['source_text']
dftest.head()
```

```python
dfval['source_text'] = "summarize: " + dfval['source_text']
dfval.head()
```

```python
from simplet5 import SimpleT5

model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
```

```python
model.train(train_df=dftrain,
            eval_df=dfval,
            outputdir="/content/drive/MyDrive/Colab Notebooks",
            batch_size=16, max_epochs=6,
            source_max_token_len=512,
            target_max_token_len=256,
            use_gpu=True)
```

```python
#! ( cd /content/drive/MyDrive/Colab Notebooks; ls )
```

```python
model.load_model("t5","/content/drive/MyDrive/Colab Notebooks/simplet5-epoch-5-train-loss-1.3971-val-loss-3.1449", use_gpu=True)
```

```python
dftest['target_text']
```

```python
reference = dftest['target_text'].tolist()
reference[0:3]
```

```python
prediction=[]
for i in dftest['source_text']:
  t5_pred=model.predict(i)
  prediction.append(t5_pred[0])
prediction[0:3]
```

```python
results_t5 = rouge.compute(predictions=prediction, references=reference)
print(results_t5) #ROUGE-1 score of 0.25 or higher is considered acceptable for scientific paper abstractive summarization
```
